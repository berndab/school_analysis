```python
# Get list of prefixes and suffixes in student names

# Student name list
student_names = student_data_df["student_name"].tolist()

# student name prefix and suffix lists
student_name_prefixes = []
student_name_suffixes = []

# Student name parts list
# Holds the student names parts after string.split()
# Prefix if present, first name, last name, suffix if present
student_name_parts = []

# For student names with name parts >= 3
# Find the prefixes and suffixes for the name
# A prefix or suffix is by position and length
# A prefix is the string student_name_parts[0] with a string length <= 4
# A suffix is the string student_name_parts[-1] with a string length <= 4
for student_name in student_names:
    
    # Spli stuent name into name parts
    student_name_parts = student_name.split()   
    
    # If a stuend name has 3 or more part
    # the student name has a prefix of a suffix
    if len(student_name_parts) >= 3:
        
        # Test for a student name prefix
        if len(student_name_parts[0]) <= 4:
            
            # Store prefix in the prefixes list
            student_name_prefixes.append(student_name_parts[0])
        
        # Test for a student name suffix
        if len(student_name_parts[-1]) <= 4:
            
            # Store the suffix in the suffixes list
            student_name_suffixes.append(student_name_parts[-1])
```

```python
# Get the unique items in the student name prefixes list.
set(student_name_prefixes)
```

```python
# Get the unique items in the studnet name suffixes list.
set(student_name_suffixes)
```

```python
# List of student name prefix and suffix to remove
prefixes_suffixes_to_remove = ["Dr. ", "Miss ", "Mr. ", "Mrs. ", "Ms. ", " DDS", " DVM", " PhD", " MD"]

# TODO only perform on student names with split() >= 3

# For each student name in the student data data fame
# Search each student name for any prefixes or suffixes in the
# prefixes_suffixes_to_remove list and remove it
for prefix_suffix in prefixes_suffixes_to_remove:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(prefix_suffix,"")
```

```python
# Verify that no student name contains a prefixes or a 
# suffix that should have been removed.

# Student name list
student_names = student_data_df["student_name"].tolist()

# Student name parts list
# Thsi will hold the student names parts after the name 
# string is split, The name parts are
# prefix(if present), first name, last name, suffix(if present)
student_name_parts = []

# list of student nammes that still have a prefix or suffix
# that should have been removed
student_names_with_prefixes_suffixes = []

# For student names with name parts >= 3, check to see if 
# the name has a prefix or suffix that should be removed
# If a student name does add it to the
# student_names_with_prefixes_suffixes list.
for student_name in student_names:

    # Split the student name into its parts
    student_name_parts = student_name.split()   
    
    # If the student name has three or more parts
    # the student name has a prefix or a suffix
    if len(student_name_parts) >= 3:
        
        # Check to see if it has a prefix of suffix that should have been removed
        if student_name_parts[0] in prefixes_suffixes_to_remove or student_name_parts[-1] in prefixes_suffixes_to_remove:
            # add the student name to the list 
            student_names_with_prefixes_suffixes.append(student_name)

# Check to see if there are any student names in the 
# student_names_with_prefixes_suffixes list
if len(student_names_with_prefixes_suffixes) == 0:
    
    # A list lenght is 0, verifys that there are no student 
    # names that contain a prefix or suffix that should have been removed
    print("SUCCESS: The student names have been cleaned of all prefixes and suffixes have beeen removed")
    
else:
    
    # A list lenght > 0, indicates that that there are are student 
    # names that contain a prefix or suffix that should have been removed.
    # The student names have NOT succefully been cleaned
    print("!!!!ERROR: The following student names still have prefixes or suffixes in the name. The student name cleaning has FAILED!!")
    print(student_names_with_prefixes_suffixes)
```
